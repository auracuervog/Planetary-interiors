```python
import numpy as np 
from matplotlib import pyplot as plt
import pylab as py
from IPython.display import Image 
```

## Maxwell construction
Given two curves for internal energy as a function of specific volume, it is possible to find a common tangent with a slope that is equal to -Pressure of phase transition. This construction is applied in this exercise.  

```python
#Volume array
v=np.linspace(0.5,2,100)

#Rough approximation of energy for molecular hydrogen
E1=(0.25)/(np.power(v,3))
#Rough approximation of energy for metallic hydrogen
E2=(1.15)+(1.1/(np.power(v,0.666)))-(1.9/(np.power(v,0.333)))

#Creating high resolution plot of the energy models for metallic and molecular H
plt.figure(dpi=100)
ax = plt.axes()
ax.grid();
 
py.plot(v,E1,label='Molecular H')
py.plot(v,E2,'g',label='Metallic H')
py.xlabel('Volume (cm3/g)')
py.ylabel('Energy (Ryd/el)')
pylab.legend(loc='upper right')


#plt.xticks(np.arange(min(v), max(v)+0.5, 0.1))
#plt.yticks(np.arange(min(E1), max(E1)+0.5, 0.1))
#plt.savefig("MaxwellConstr-H.tiff",dpi=150)

```

The circle that intersects the common tangent and the model for metallic hydrogen has coordinates (x2,y2), where x2=[0.57;0.62] and y2=[0.40;0.50]. The ellipse in the point of intersection between the common tangent and the model for molecular hydrogen has coordinates (x1,y1), where x1=[1.07;1.16] and y1=[0.13;0.23]. Therefore, I calculate the average values for the coordinates of the intersections with the tangent to find its slope in the following lines of code:

```python
x2min=0.57
x2max=0.62
y2min=0.4
y2max=0.5

x2av=(x2min+x2max)/2
y2av=(y2min+y2max)/2

x1min=1.07
x1max=1.16
y1min=0.13
y1max=0.23

x1av=(x1min+x1max)/2
y1av=(y1min+y1max)/2

#Mean slope of the common tangent
m=(y2av-y1av)/(x2av-x1av)
#Minimun slope of the common tangent
mmi=(y2min-y1min)/(x2min-x1min)
#Maximun slope of the common tangent
max=(y2max-y1max)/(x2max-x1max)

print("The mean slope of the common tangent is {}".format(m))
print("The minimun slope of the common tangent is {}".format(mmi))
print("The maximum slope of the common tangent is {}".format(max))
```

From the graph, the y-intercept is ~0.53. Therefore, an approximate equation for the tangent common to both energy models is: y = mx + b -> Mean E_predicted = -0.52v + 0.53; Minimum E_predicted = -0.54v + 0.53; Maximum E_predicted = -0.5v + 0.53.

```python
MEpre=(-0.52*v)+0.53
MiEpre=(-0.54*v)+0.53
MaEpre=(-0.5*v)+0.53

plt.figure(dpi=100)
ax = plt.axes()
ax.grid();
 
py.plot(v,E1,label='Molecular H')
py.plot(v,E2,'g',label='Metallic H')
py.plot(v,MEpre,label='Mean common tangent')
py.plot(v,MiEpre,'b',label='Minimum common tangent')
py.plot(v,MaEpre,'b',label='Maximum common tangent')
py.xlabel('Volume (cm3/g)')
py.ylabel('Energy (Ryd/el)')
pylab.legend(loc='upper right')
```
